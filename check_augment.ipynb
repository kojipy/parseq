```python
import torchvision.transforms as T
import hydra
from omegaconf import OmegaConf

from strhub.data.module import AbgalDataModule, KeepAspectResize, Pad, ImageCompression
```

```python
config = OmegaConf.load("configs/main.yaml")
data_module = hydra.utils.instantiate(config.data)
transforms = AbgalDataModule.get_transform(augment=True, img_width=config.data.img_width, img_height=config.data.img_height)
train_dataset = data_module.train_dataset
```

```python
img_width = config.data.img_width
img_height = config.data.img_height

transforms = T.Compose([
    KeepAspectResize((img_width, img_height)),
    ImageCompression(quality_upper=2),
    Pad((img_width, img_height)),
    T.RandomApply(
        transforms=[T.GaussianBlur(kernel_size=(3, 5), sigma=(0.1, 1))],
        p=0.3,
    ),
    T.RandomRotation(degrees=(0, 1)),
    T.Resize(img_height), T.Grayscale(), T.ToTensor()
])
```
